```python
from bs4 import BeautifulSoup
import requests as req
import urllib.request
import os
```

```python
resp = req.get("https://www.kompas.com/tren/read/2023/10/02/060000965/sejarah-hari-batik-nasional-dan-cara-merayakannya?page=all")
soup = BeautifulSoup(resp.text,'lxml')
```

```python
for x in soup.find_all('h1'):
    print(x.text)
```

```python
for x in soup.find_all('p'):
    print(x.text)
```

```python
for x in soup.find_all('h1',{"class":"read__title"}):
    if "Batik" in x.text:
        print(x.text)
```

```python
for a in soup.find_all('a',href=True):
    if("http" in a['href']):
        print(a['href'])
```

```python
f = open("getLink.txt",'w')
for a in soup.find_all('a',href=True):
    if("https" in a['href']):
        f.write(a['href']+"\n")
```

```python
for img in soup.find_all('img',src=True):
    print ("https:"+img['src'])
```

```python
fileformat = [".jpg",".jpeg",".JPG",".png"]
for img in soup.find_all('img', src=True):
    img_data = img['src'].split("/")
    
    if any(x in img_data[-1] for x in [".jpg", ".jpeg", ".JPG", ".png"]):
        try:
            img_name = img_data[-1]
            urllib.request.urlretrieve(img['src'], os.path.join("img", img_name))
            print(img_name + " sukses")
        except Exception as e:
            print(str(e))
            print(img_name + " gagal")
```
